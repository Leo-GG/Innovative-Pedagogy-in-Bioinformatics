```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df_sc=pd.read_excel('data/Python_Fundamentals.xlsx')
df_sc['Group']='2'
```

```python
df_bhs=pd.read_excel('BHS.xlsx')
df_bhs['Group']='1'
```

```python
df=pd.concat([df_bhs, df_sc])
```

```python
df.columns=[i.split('\n')[0] for i in df.columns]
```

```python
df_melted=df.iloc[:,6:].melt(ignore_index=False, value_vars=df.columns[6:-1],id_vars='Group')
```
